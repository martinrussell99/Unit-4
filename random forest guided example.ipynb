```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Replace the path with the correct path for your data.
y2015 = pd.read_csv(r'C:\Users\Martin\Desktop\LoanStats3d.csv',
    skipinitialspace=True,
    header=1)
# Note the warning about dtypes.
```

```python
# Convert ID and Interest Rate to numeric.
y2015['id'] = pd.to_numeric(y2015['id'], errors='coerce')
y2015['int_rate'] = pd.to_numeric(y2015['int_rate'].str.strip('%'), errors='coerce')
print(y2015.shape)

# Drop other columns with many unique variables
y2015.drop(['url', 'emp_title', 'zip_code', 'earliest_cr_line', 'revol_util',
            'sub_grade', 'addr_state', 'desc', 'id', 'funded_amnt', 'funded_amnt_inv', 'installment', 'policy_code',
           'hardship_length', 'deferral_term', 'revol_bal_joint', 'hardship_dpd', 'out_prncp_inv'], 1, inplace=True)
y2015 = y2015[y2015.columns.drop(list(y2015.filter(regex='sec_app')))]
y2015 = y2015[y2015.columns.drop(list(y2015.filter(regex='mths')))]
y2015 = y2015[y2015.columns.drop(list(y2015.filter(regex='recover')))]

print(y2015.shape)
```

```python
y2015 = y2015[:-4]
```

```python
print(y2015.shape)
```

```python
means = y2015[['total_rec_int', 'orig_projected_additional_accrued_interest', 'hardship_payoff_balance_amount']].mean(axis=0)
stds = y2015[['total_rec_int', 'orig_projected_additional_accrued_interest', 'hardship_payoff_balance_amount']].std(axis=0)
y2015['eng1'] = ((y2015[['total_rec_int', 'orig_projected_additional_accrued_interest', 'hardship_payoff_balance_amount']] - means) / stds).mean(axis=1)
means = y2015[['num_sats', 'num_op_rev_tl', 'open_acc']].mean(axis=0)
stds = y2015[['num_sats', 'num_op_rev_tl', 'open_acc']].std(axis=0)
y2015['eng2'] = ((y2015[['num_sats', 'num_op_rev_tl', 'open_acc']] - means) / stds).mean(axis=1)
means = y2015[['total_pymnt', 'total_pymnt_inv', 'total_rec_prncp']].mean(axis=0)
stds = y2015[['total_pymnt', 'total_pymnt_inv', 'total_rec_prncp']].std(axis=0)
y2015['eng3'] = ((y2015[['total_pymnt', 'total_pymnt_inv', 'total_rec_prncp']] - means) / stds).mean(axis=1)
means = y2015[['settlement_amount', 'hardship_amount']].mean(axis=0)
stds = y2015[['settlement_amount', 'hardship_amount']].std(axis=0)
y2015['eng4'] = ((y2015[['settlement_amount', 'hardship_amount']] - means) / stds).mean(axis=1)


y2015 = y2015.drop(['total_rec_int', 'orig_projected_additional_accrued_interest', 'hardship_payoff_balance_amount',
                   'num_sats', 'num_op_rev_tl', 'open_acc','total_pymnt', 'total_pymnt_inv', 'total_rec_prncp',
                   'settlement_amount', 'hardship_amount'], 1)
```

```python
print(y2015.shape)
```

```python
pd.get_dummies(y2015)
```

```python
from sklearn import ensemble
from sklearn.model_selection import cross_val_score

rfc = ensemble.RandomForestClassifier()
X = y2015.drop('loan_status', 1)
Y = y2015['loan_status']
X = pd.get_dummies(X)

```

```python
X = X.dropna(axis=1)
```

```python
print(y2015.shape)
print(X.shape)
print(Y.shape)
```

```python
cross_val_score(rfc, X, Y, cv=10)
```
